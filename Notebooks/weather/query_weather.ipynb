```python
import pandas as pd
import numpy as np
from datetime import datetime
from meteostat import Stations, Hourly
from meteostat.units import fahrenheit, direction, condition
from meteostat import Monthly
from meteostat import Point, Daily
import matplotlib.pyplot as plt
```

```python
df_position_station = pd.read_csv("/content/edited_db.csv")
df_position_station['time'] = pd.to_datetime(df_position_station['date_str'])
print(df_position_station.shape)
df_position_station.head()
```

```python
position_features = ['Latitude', 'Longitude', 'Altitude']
position = df_position_station[position_features]
example_pos = tuple(position.iloc[1])
```

```python
# Create columns if they do not exist.
for col in ['tavg', 'prcp', 'wspd', 'pres', 'tsun']:
    if col not in df_position_station.columns:
        df_position_station[col] = np.nan

df_position_station.head()
```

```python
for i in range(len(df_position_station)):
  try:
    # Set time period
    start = df_position_station['time'].iloc[i]

    position = (df_position_station['Latitude'].iloc[i], df_position_station['Longitude'].iloc[i])

    # Get Hourly data
    stations = Stations()
    stations = stations.nearby(position[0], position[1])
    station = stations.fetch(1)

    station_id = station.index[0]
    date_tuple = (start.year, start.month, start.day)

    start = datetime(date_tuple[0], date_tuple[1], date_tuple[2])
    end = start

    # Get Monthly data
    data = Monthly(station_id, start, end)
    data = data.fetch()

    if not data.empty:
        df_position_station.at[i, 'tavg'] = data['tavg'].iloc[0]
        df_position_station.at[i, 'prcp'] = data['prcp'].iloc[0]
        df_position_station.at[i, 'wspd'] = data['wspd'].iloc[0]
        df_position_station.at[i, 'pres'] = data['pres'].iloc[0]
        df_position_station.at[i, 'tsun'] = data['tsun'].iloc[0]
    else:
        df_position_station.at[i, 'tavg'] = np.nan
        df_position_station.at[i, 'prcp'] = np.nan
        df_position_station.at[i, 'wspd'] = np.nan
        df_position_station.at[i, 'pres'] = np.nan
        df_position_station.at[i, 'tsun'] = np.nan
  except:
    continue
```

```python
# Count null values in each column and print the shape of the DataFrame
null_counts = df_position_station.isnull().sum()
print(null_counts)
df_position_station.shape
```

```python
# Remove rows with any null values
df_position_station = df_position_station.dropna()

# Print the shape of the DataFrame after removing null values
df_position_station.shape
```

```python
df_position_station.to_csv("example_edited_db_full.csv", index=False)
```
